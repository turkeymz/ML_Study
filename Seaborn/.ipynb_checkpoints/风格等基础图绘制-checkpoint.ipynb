```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
# 写完代码直接在Jupyter显示图
%matplotlib inline
```

```python
def sinplot(flip=1):
    # 0,14找100个点
    x = np.linspace(0,14,100)
    # 画6条线
    for i in range(1,7):
        plt.plot(x,np.sin(x + i * 5) * (7 - i) * flip)
```

```python
sinplot()
```

```python
# 使用Seaborn默认参数
sns.set()
sinplot()
```

```python
# 设置白格子风格
sns.set_style('whitegrid')
data = np.random.normal(size=(20,6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

```python
# 设置暗模式无刻度线风格
sns.set_style('dark')
sinplot()
```

```python
# 设置白模式无刻度线风格
sns.set_style('white')
sinplot()
```

```python
# 设置白模式有刻度线风格
sns.set_style('ticks')
sinplot()
```

```python
# 设置白模式有刻度线风格
sinplot()
# 去掉无用的线框
sns.despine()
```

```python
sns.violinplot(data)
sns.despine(offset=10)
```
