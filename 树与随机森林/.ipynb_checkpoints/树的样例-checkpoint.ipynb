```python
import pandas as pd
import matplotlib.pyplot as plt
# 读取数据集
iris_df = pd.read_csv('iris.data')
iris_df.columns = ['sepal_length_cm','sepal_width_cm','petal_length_cm','petal_width_cm','class']
iris_df.head()
```

```python
from PIL import Image
img = Image.open('test.jpg')
plt.imshow(img)
plt.show()
```

```python
# 查看花每一个特征的统计值
iris_df.describe()
```

```python
import seaborn as sb
# 绘制矩阵内每个特征两两的散点图，必须不能有缺失值，所以要dro
sb.pairplot(iris_df.dropna(),hue='class')
```

```python
from sklearn.tree import DecisionTreeClassifier
print(help(DecisionTreeClassifier))
```

```python
from sklearn.cross_validation import train_test_split
all_inputs = iris_df[['sepal_length_cm','sepal_width_cm','petal_length_cm','petal_width_cm']].values
all_classes = iris_df['class'].values
(training_inputs,test_inputs,training_classes,test_classes) = train_test_split(all_inputs,all_classes,train_size=0.75,random_state=1)
print(training_inputs)
```

```python
decision_tree_classifier = DecisionTreeClassifier()
# 构建决策树
decision_tree_classifier.fit(training_inputs,training_classes)
#输出测试数据与标签的准确度
decision_tree_classifier.score(test_inputs,test_classes)
```
