```python
import pandas as pd
import numpy as np
train_df = pd.read_csv('titanic_train.csv')
train_df.head(10)
```

```python
age = train_df['Age']
# 判断缺失值
age_is_null = pd.isnull(age)
#print(age_is_null)
# 获取缺失值的行
age_null_true = age[age_is_null]
print(age_null_true)
# 获取缺失值的长度
print(len(age_null_true))
```

```python
#求age的非缺失值
good_ages = train_df['Age'][age_is_null == False]
# 求age的均值
print(train_df['Age'].mean())
```

```python
# 以船舱Pclass为索引，统计获救率Survived，方法是np.mean
passenger_survival = train_df.pivot_table(index='Pclass',values='Survived',aggfunc=np.mean)
print(passenger_survival)
```

```python
drop_na_columns = train_df.dropna(axis=1)
print(drop_na_columns)
```
